```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import PIL   #Image Library in Python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

# 2.Convert Images into Pixels and store in x,y

```python
flowers_image_dict = {
    'roses':list(data_dir.glob('roses/*')),
    'tulips': list(data_dir.glob('tulips/*')),
    'sunflowers': list(data_dir.glob('sunflowers/*')),
    'dandelion': list(data_dir.glob('dandelion/*')),
    'daisy': list(data_dir.glob('daisy/*'))
}
flowers_label_dict ={
    'roses':0,'daisy':1,'dandelion':2,'sunflowers':3,'tulips':4
}
flowers_image_dict['roses'];
```

```python
flowers_image_dict['roses'][0]
#first image path
```

```python
#read the image files into OPENCV object (sort of numpy array)
img = cv2.imread(str(flowers_image_dict['roses'][0]))
img.shape
```

```python
#making same dimesions for all the images
cv2.resize(img,(180,180)).shape
```

```python
#run for loop and prepare x and y
x,y = [],[]
for flower_name,images in flowers_image_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_image = cv2.resize(img,(180,180))
        x.append(resized_image)
        y.append(flowers_label_dict[flower_name])
```

```python
for flower_name,images in flowers_image_dict.items():
    print(flower_name)
    print(len(images))
```

```python
x[0]; #3d x,y, and rgb 
```

```python
x =  np.array(x)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
len(x_train),len(x_test)
```

```python
np.max(x)
```

```python
x_train_scaled = x_train/255
x_test_scaled = x_test/255
```

# 3. CNN Model

```python
model = Sequential([
    #cnn
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
    #ann
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(5)
])
```

```python
model.compile(optimizer='adam',
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model.fit(x_train_scaled,y_train,epochs=30)
```

```python
model.evaluate(x_test_scaled,y_test)
```

# OVERFITTING ^|

```python
predictions = model.predict(x_test_scaled)
score = tf.nn.softmax(predictions[2])
np.argmax(score),y_test[2]
```

# 4.Data Augmentation

```python
data_augmentation = keras.Sequential([
     #A preprocessing layer which randomly zooms images during training.
    #layers.experimental.preprocessing.RandomZoom(0.9)
    #layers.experimental.preprocessing.RandomContrast(0.5)
    #layers.experimental.preprocessing.RandomRotation(0.3)
    layers.RandomFlip("horizontal",input_shape=(180,180,3)),
      layers.RandomRotation(0.1),
      layers.RandomZoom(0.1)
])
```

```python
plt.axis('off') #to remove wierd scale numbers on axes
plt.imshow(x[0])
```

```python
plt.axis('off')
plt.imshow(data_augmentation(x)[0].numpy().astype('uint8'))
```

```python
model2 = Sequential([
  data_augmentation,#####
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.2),
    #ann
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(5)
])
```

```python
model2.compile(optimizer='adam',
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
model2.fit(x_train_scaled,y_train,epochs=30)
```

```python
model2.evaluate(x_test_scaled,y_test)
```
