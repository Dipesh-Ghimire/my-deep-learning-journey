```python
import pandas as pd
pd.options.mode.chained_assignment = None
import tensorflow
from tensorflow import keras
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report

df = pd.read_csv('customer_churn_Preprocessed_data.csv',index_col=0)
```

```python
def ANN(x_train_ANN,x_test_ANN,y_train_ANN,y_test_ANN,loss_ANN,weights_ANN):
    model_ANN = keras.Sequential([
    keras.Input(shape=(26,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')])
    
    model_ANN.compile(optimizer='adam',loss=loss_ANN,metrics=['accuracy'])
    if weights_ANN==-1:
        model_ANN.fit(x_train_ANN,y_train_ANN,epochs=100)
    else:
        model_ANN.fit(x_train_ANN,y_train_ANN,epochs=100,class_weight=weights_ANN)
    print(model_ANN.evaluate(x_test_ANN,y_test_ANN))
    y_preds_ANN = model_ANN.predict(x_test_ANN)
    y_preds_ANN = np.round(y_preds_ANN)
    print(classification_report(y_test_ANN,y_preds_ANN))
    return y_preds_ANN
```

```python
x = df.drop('Churn',axis='columns')
y = df['Churn']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
count_class_0,count_class_1 = y_train.value_counts()
y_train.value_counts()
#imbalanced
```

```python
df_train = pd.concat([pd.DataFrame(x_train),pd.DataFrame(y_train)],axis='columns')

df_train_class_0 = df_train[df_train.Churn==0]
df_train_class_1 = df_train[df_train.Churn==1]

df_class_1_over = df_train_class_1.sample(count_class_0,replace=True)

new_df_train = pd.concat([df_train_class_0,df_class_1_over],axis=0)

new_x_train = new_df_train.drop('Churn',axis='columns')
new_y_train = new_df_train.Churn
```

```python
y_predicted = ANN(new_x_train,x_test,new_y_train,y_test,'binary_crossentropy',-1)
```
