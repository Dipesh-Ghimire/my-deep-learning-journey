```python
import pandas as pd
pd.options.mode.chained_assignment = None
import tensorflow
from tensorflow import keras
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report

df = pd.read_csv('customer_churn_Preprocessed_data.csv',index_col=0)
```

```python
def ANN(x_train_ANN,x_test_ANN,y_train_ANN,y_test_ANN,loss_ANN,weights_ANN):
    model_ANN = keras.Sequential([
    keras.Input(shape=(26,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')])
    
    model_ANN.compile(optimizer='adam',loss=loss_ANN,metrics=['accuracy'])
    if weights_ANN==-1:
        model_ANN.fit(x_train_ANN,y_train_ANN,epochs=100)
    else:
        model_ANN.fit(x_train_ANN,y_train_ANN,epochs=100,class_weight=weights_ANN)
    print(model_ANN.evaluate(x_test_ANN,y_test_ANN))
    y_preds_ANN = model_ANN.predict(x_test_ANN)
    y_preds_ANN = np.round(y_preds_ANN)
    print(classification_report(y_test_ANN,y_preds_ANN))
    return y_preds_ANN
```

```python
x = df.drop('Churn',axis='columns')
y = df['Churn']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
count_class_0,count_class_1 = y.value_counts()
y_train.value_counts()
#imbalanced
```

```python
df2 = x_train.copy()
df2['Churn'] = y_train
```

```python
df2_class0 = df2[df2.Churn==0]
df2_class1 = df2[df2.Churn==1]
df2_class0.shape,df2_class1.shape
```

```python
def get_train_batch(df_majority,df_minority,start,end):
    df_train = pd.concat([df_majority[start:end],df_minority],axis=0)
    x_train = df_train.drop('Churn',axis = 'columns')
    y_train = df_train.Churn
    return x_train,y_train
```

```python
x_train_1,y_train_1 =get_train_batch(df2_class0,df2_class1,0,1502)
y_pred_1 = ANN(x_train_1,x_test,y_train_1,y_test,'binary_crossentropy',-1)
```

```python
x_train_2,y_train_2 =get_train_batch(df2_class0,df2_class1,1502,3004)
y_pred_2 = ANN(x_train_2,x_test,y_train_2,y_test,'binary_crossentropy',-1)
```

```python
x_train_3,y_train_3 =get_train_batch(df2_class0,df2_class1,3004,4123)
y_pred_3 = ANN(x_train_3,x_test,y_train_3,y_test,'binary_crossentropy',-1)
```

```python
y_pred_final = y_pred_1.copy()
for i in range(len(y_pred_1)):
    sum = y_pred_1[i]+y_pred_2[i]+y_pred_3[i]
    if sum > 1:
        y_pred_final[i]=1
    else:
        y_pred_final[i]=0
```

```python
print(classification_report(y_test,y_pred_final))
```
